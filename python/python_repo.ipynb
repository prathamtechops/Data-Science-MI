```python
import requests 
```

```python
url = 'https://api.github.com/search/repositories?q=language:python&sort=stars'
r = requests.get(url)
```

```python
print("Status Code: ", r.status_code)
```

```python
response_dict = r.json()
response_dict.keys()
```

```python
print("Total repositories ", response_dict['total_count'])
```

```python
repo_dicts = response_dict['items']
repo_dicts
```

```python
print("Keys" , len(repo_dicts))
```

```python
print("Selected information about the first repository: ")
for repo_dict in repo_dicts:
    print("\nName: ", repo_dict['name'])
    print("Owner: ", repo_dict["owner"]['login'] )
    print("Stars: ", repo_dict['stargazers_count'])
    print("Repository: ", repo_dict['html_url'])
    print("Description: ", repo_dict['description'])
    
```

```python
import plotly.graph_objects as go
```

```python
names = [repo['name'] for repo in repo_dicts]
stars = [repo['stargazers_count'] for repo in repo_dicts]
urls = [repo['html_url'] for repo in repo_dicts]
descriptions = [repo['description'] for repo in repo_dicts]
```

```python


# Create the interactive bar chart
fig = go.Figure()

# Add the bar trace
bar_trace = go.Bar(
    x=names,
    y=stars,
    hovertext=stars,
    text=descriptions,
    marker_color='lightskyblue',
    textposition='outside'
)

# Update tooltip behavior
bar_trace.hoverinfo = 'x+y+text'  # Show repository name, stars, and description in tooltip

# Add the bar trace to the figure
fig.add_trace(bar_trace)

# Update layout to add axis and title
fig.update_layout(
    title='Most Starred Repositories',
    xaxis_title='Repository',
    yaxis_title='Stars',
    xaxis_tickangle=-45
)

# Show the interactive bar chart
fig.show()
```
