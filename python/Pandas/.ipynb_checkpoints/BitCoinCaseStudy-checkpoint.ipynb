# üîç Bitcoin Price Analysis
## Dataset: Yahoo Finance Bitcoin Historical Data (2014-2023)

```python
import pandas as pd
df = pd.read_csv("BTC-USD.csv")
df
```

The price of Bitcoin goes up and down based on how many people want to buy it (demand) and how much of it is available (supply). If more people want to buy Bitcoin, the price goes up. If fewer people want to buy it, the price goes down. The price can also be affected by news, like good or bad things happening in the world, and how many people are using Bitcoin for things like buying stuff or investing.

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.set_index('Date', inplace=True)
df
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

# Line plot of Bitcoin's closing price over time
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Close'], label='Bitcoin Price', color='b')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.title('Bitcoin Price Analysis')
plt.legend()
plt.show()
```

Bitcoin reached its all-time high price on November 10, 2021, when it was valued at approximately $68,958 per coin. That particular price hike was connected to the launch of a Bitcoin ETF in the United States, while others in 2021 were due to events involving Tesla and Coinbase, respectively1. However, by the end of 2022, the market was noticeably different, with Bitcoin prices reaching roughly 29,848.03 as of July 19, 2023 after another crypto exchange, FTX, filed for bankruptcy1.

```python
# Calculate 50-day and 200-day moving averages
df['MA_50'] = df['Close'].rolling(window=50).mean()
df['MA_200'] = df['Close'].rolling(window=200).mean()

# Plot Bitcoin price and moving averages
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['MA_50'], label='50-day MA', color='g', linestyle='--')
plt.plot(df.index, df['MA_200'], label='200-day MA', color='r', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.title('Bitcoin Price Analysis with Moving Averages')
plt.legend()
plt.show()
```

The Covid-19 pandemic and the Federal Reserve‚Äôs stimulus measures in 2020: The pandemic caused a global economic downturn, leading central banks to implement stimulus measures such as quantitative easing. This led to concerns about inflation and currency devaluation, which boosted the demand for Bitcoin as a store of value.
The interest rate hikes and the stablecoin collapse in 2021-2023: In response to rising inflation, central banks began to raise interest rates. This led to a collapse of several stablecoins, which were pegged to fiat currencies. This caused a loss of confidence in stablecoins and a flight to safety into Bitcoin.
