```python
import pandas as pd
```

```python
reaction_df = pd.read_csv('Reactions.csv')
reaction_type_df = pd.read_csv('ReactionTypes.csv')
```

```python
reaction_df = reaction_df.dropna()
reaction_type_df
```

```python
merged_df = pd.merge(reaction_df, reaction_type_df, on='Type', how='inner')  # You can change 'how' to 'left', 'right', or 'outer' as needed
```

```python
merged_df
```

```python
merged_df.to_csv('merged_file.csv', index=False)
```

```python
content_df = pd.read_csv("Content.csv")
```

```python
content_df = content_df.dropna()
```

```python
final_merged_df = pd.merge(new_df, content_df, on='Content ID', how='inner')  
final_merged_df.to_csv('final_merged_file.csv', index=False)
```

```python
final_merged_df
```

```python
merged_df = pd.read_csv('final_merged_file.csv')

# Group by 'Category' and sum up the 'Score' column
category_scores = merged_df.groupby('Category')['Score'].sum()

# Sort the results in descending order to find the top performing categories
top_categories = category_scores.sort_values(ascending=False).head(5)

# Convert the Series to a DataFrame
top_categories_df = top_categories.reset_index()
top_categories_df.columns = ['Category', 'Total_Score']

# Save the DataFrame to a CSV file
top_categories_df.to_csv('top_categories.csv', index=False)
```

```python
data = pd.read_csv('final.csv')

# Group by 'Category' and sum the 'Score'
category_scores = data.groupby('Category')['Score'].sum()

# Find the top 5 categories
top_categories = category_scores.sort_values(ascending=False).head(5)

print(top_categories)
```

```python
import matplotlib.pyplot as plt

# Sample data (replace with your data)
data = {
    'Category': ['animals', 'science', 'healthy eating', 'technology', 'food'],
    'Score': [74965, 71168, 69339, 68738, 66676]
}

# Create a DataFrame from the sample data
df = pd.DataFrame(data)

# Create a pie chart
plt.figure(figsize=(10, 6))
plt.pie(df['Score'], labels=df['Category'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Content Categories (Pie Chart)')
plt.savefig('pie_chart.png')  # Save the pie chart as an image
plt.show()

# Create a bar graph
plt.figure(figsize=(10, 6))
plt.bar(df['Category'], df['Score'])
plt.title('Top 5 Content Categories (Bar Graph)')
plt.xlabel('Category')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('bar_graph.png')  # Save the bar graph as an image
plt.show()
```
